## Narrative

This lab uses the KDD Cup network dataset as a security-analytics exercise. Each network connection is treated as an observation. K-means groups similar connections into clusters, allowing unusually small, distant, or high-error clusters to be reviewed as possible anomalies.

This is not a full intrusion-detection system. It is an exploratory workflow for feature parsing, behavior grouping, centroid interpretation, and anomaly triage.


## Use Cases

- Network anomaly detection
- SOC enrichment experiments
- Baseline behavior modeling
- Unsupervised clustering of large records

## Step 4 — Interpret Clusters as Security Signals

Centroids summarize typical behavior in each cluster. Large clusters often represent common traffic patterns. Small clusters, high-distance points, or clusters dominated by unusual labels can become candidates for deeper analysis.
